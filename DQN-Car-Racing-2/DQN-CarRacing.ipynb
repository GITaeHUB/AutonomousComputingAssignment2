```python
env = gym.make('CarRacing-v2', continuous=False)
env = ImageEnv(env)

max_steps = int(9e4)
eval_interval = 100
state_dim = (4, 84, 84)
action_dim = env.action_space.n

agent = DQN.DQN(state_dim, action_dim)
```

```python
def evaluate(n_evals=5):
    eval_env = gym.make('CarRacing-v2', continuous=False)
    eval_env = ImageEnv(eval_env)
    
    scores = 0
    for i in range(n_evals):
        (s, _), done, ret = eval_env.reset(), False, 0
        while not done:
            a = agent.act(s, training=False)
            s_prime, r, terminated, truncated, info = eval_env.step(a)
            s = s_prime
            ret += r
            done = terminated or truncated
        scores += ret
    return np.round(scores / n_evals, 4)
```

```python
import os
import torch
from IPython.display import clear_output

# 기존에 저장된 모델과 상태가 있는지 확인
checkpoint_path = 'dqn_checkpoint.pt'
history_path = 'training_history.pt'
start_step = 0
history = {'Step': [], 'AvgReturn': []}

# 모델과 로그 불러오기
if os.path.exists(checkpoint_path):
    checkpoint = torch.load(checkpoint_path)
    agent.network.load_state_dict(checkpoint['network_state_dict'])
    agent.target_network.load_state_dict(checkpoint['target_network_state_dict'])
    agent.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    start_step = checkpoint['total_steps']
    
# 로그 불러오기 (기존 로그가 있을 경우 이어서 기록)
if os.path.exists(history_path):
    history = torch.load(history_path)

# 훈련 시작
(s, _) = env.reset()
agent.total_steps = start_step  # 이어서 시작하기 위해 초기화
while True:
    a = agent.act(s)
    s_prime, r, terminated, truncated, info = env.step(a)
    result = agent.process((s, a, r, s_prime, terminated))
    
    s = s_prime
    if terminated or truncated:
        s, _ = env.reset()
        
    # 평가 및 기록 저장
    if agent.total_steps % eval_interval == 0:
        ret = evaluate()
        history['Step'].append(agent.total_steps)
        history['AvgReturn'].append(ret)
        
        clear_output()
        plt.figure(figsize=(8, 5))
        plt.plot(history['Step'], history['AvgReturn'], 'r-')
        plt.xlabel('Step', fontsize=16)
        plt.ylabel('AvgReturn', fontsize=16)
        plt.xticks(fontsize=14)
        plt.yticks(fontsize=14)
        plt.grid(axis='y')
        plt.show()
        
        # 모델과 학습 상태 저장
        torch.save({
            'network_state_dict': agent.network.state_dict(),
            'target_network_state_dict': agent.target_network.state_dict(),
            'optimizer_state_dict': agent.optimizer.state_dict(),
            'total_steps': agent.total_steps
        }, checkpoint_path)

        # 로그 저장
        torch.save(history, history_path)
    
    if agent.total_steps > max_steps:
        break
```

```python
eval_env = gym.make('CarRacing-v2', continuous=False, render_mode='rgb_array')
eval_env = ImageEnv(eval_env)

frames = []
scores = 0
(s, _), done, ret = eval_env.reset(), False, 0
while not done:
    frames.append(eval_env.render())
    a = agent.act(s, training=False)
    s_prime, r, terminated, truncated, info = eval_env.step(a)
    s = s_prime
    ret += r
    done = terminated or truncated
scores += ret
```
